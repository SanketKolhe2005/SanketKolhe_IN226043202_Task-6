#  Problem Statement 1:Smart Parking Lot Management System

```python
def smart_parking_system(capacity, vehicle_logs):
    current_count = 0
    peak_usage = 0
    
    # Loop through logs
    for log in vehicle_logs:
        if log == "IN":
            current_count += 1
        elif log == "OUT":
            current_count -= 1
        
        # Track peak usage
        if current_count > peak_usage:
            peak_usage = current_count
    
    # Parking status
    if current_count > capacity:
        status = "Over Capacity"
    else:
        status = "Available"
    
    print("Currently Parked Vehicles:", current_count)
    print("Peak Usage:", peak_usage)
    print("Parking Status:", status)


# Sample Test
smart_parking_system(50, ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"])
```

#  Problem Statement 2:Online Food Delivery Time Estimator

```python
def delivery_time_estimator(distance, traffic, weather):
    base_time = distance * 5  # 5 minutes per km
    
    # Traffic adjustment
    if traffic == "High":
        base_time += 10
    elif traffic == "Medium":
        base_time += 5
    
    # Weather adjustment
    if weather == "Rainy":
        base_time += 5
    elif weather == "Storm":
        base_time += 15
    
    print("Estimated Delivery Time:", base_time, "minutes")


# Sample Test
delivery_time_estimator(8, "High", "Rainy")
```

#  Problem Statement 3:Movie Theatre Seat Occupancy Analyzer

```python
def theatre_occupancy(total_seats, booked_seats):
    booked_count = len(booked_seats)
    occupancy_percentage = (booked_count / total_seats) * 100
    
    # Determine show status
    if occupancy_percentage == 100:
        status = "Housefull"
    elif occupancy_percentage >= 75:
        status = "Almost Full"
    else:
        status = "Seats Available"
    
    print("Occupancy:", round(occupancy_percentage), "%")
    print("Show Status:", status)


# Sample Test
theatre_occupancy(200, [1]*150)
```

# Problem Statement 4:Cloud Server Load Classification System

```python
def server_load_classifier(cpu_readings):
    average_load = sum(cpu_readings) / len(cpu_readings)
    
    # Classification rules
    if average_load < 50:
        status = "Normal"
    elif 50 <= average_load <= 80:
        status = "Warning"
    else:
        status = "Critical"
    
    print("Average CPU Load:", round(average_load), "%")
    print("Server Status:", status)


# Sample Test
server_load_classifier([45, 60, 70, 85, 90])
```

# Problem Statement 5:Smart Classroom Resource Usage Monitor

```python
def classroom_resource_monitor(resource_usage):
    overused = []
    
    # Assume overuse threshold = 8 hours
    for resource, hours in resource_usage.items():
        if hours > 8:
            overused.append(resource)
    
    if overused:
        energy_alert = "Yes"
    else:
        energy_alert = "No"
    
    print("Overused Resources:", ", ".join(overused))
    print("Energy Alert:", energy_alert)


# Sample Test
resources = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

classroom_resource_monitor(resources)
```
